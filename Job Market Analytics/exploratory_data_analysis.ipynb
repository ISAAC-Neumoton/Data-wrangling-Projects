```python
top_jobs.plot(kind = 'barh')
```

```python
# Which job titles have the most missing salary information?

# Calculate missing yearly salary by job title
missing_yearly = df[df['salary_year_avg'].isna()].groupby('job_title_short').size().reset_index(name='missing_year_count')

# Add total count for percentage calculation
total_by_title = df.groupby('job_title_short').size().reset_index(name='total_count')

# Merge and calculate percentage
yearly_analysis = pd.merge(missing_yearly, total_by_title, on='job_title_short')
yearly_analysis['missing_year_pct'] = (yearly_analysis['missing_year_count'] / yearly_analysis['total_count'] * 100)
yearly_analysis['missing_hour_pct'] = (100- yearly_analysis['missing_year_count'] / yearly_analysis['total_count'] * 100)

# Sort by count of missing yearly salaries
yearly_analysis = yearly_analysis.sort_values('missing_year_count', ascending=False)

yearly_analysis.head(10)
```

```python
import numpy as np

# Create pay_per_hr
df["pay_per_hr"] = np.where(
    df["salary_hour_avg"].notna(),
    df["salary_hour_avg"],
    df["salary_year_avg"] / 2080
)

# Create pay_per_annum
df["pay_per_annum"] = np.where(
    df["salary_year_avg"].notna(),
    df["salary_year_avg"],
    df["salary_hour_avg"] * 2080
)

# Drop the old column
```
